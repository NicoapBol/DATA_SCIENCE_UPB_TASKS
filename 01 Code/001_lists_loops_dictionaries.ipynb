### 00. Listas

```python
mult_de_5 = [5,10,15,20,25]
b = mult_de_5
print("Tenemos dos listas mult_de_5 y b que es su clon, el principio de 'Aliasing' significa tener varias objetos que refieren a una mismo objeto, lo que modifica una de las listas afecta a la otra")
print("+"+"="*60*4+"+")
print(f"mult_de_5 = {mult_de_5}")
print(f"b = {b}")
print("+"+"="*60*4+"+")
print("Retirar el ultimo elemnto con .pop()")
mult_de_5.pop()
print(f"mult_de_5 = {mult_de_5}")
print(f"b = {b}")
print("+"+"="*60*4+"+")
print("Evidentemente lo que modifico en mult_de_5 afecta a b")
```
